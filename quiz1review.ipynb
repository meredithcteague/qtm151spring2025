LECTURE 7: user defined functions 
- there are many python functions that are built in, such as len(), abs(), print(), type(), round().
- every function has parameters, arguments and returns. parameter: variable in the function. argument: actual value that is given to a parameter depending on the function call. return: values that the function returns after executing. 
- how to define your own function: 
- def functionname(parameter1, parameter2):

- generally: function name = lambda, parameter1, parameter2: method body. you call it the same way as a regular function. 


LECTURE 8: defining functions and global local variables
- global variable: variable that has a value stored outside of a function -> you can see them in environment, and you can use them outside of the code block that they were defined in. you can use them in function, but it is recommended that you put things in parameters instead so it is less confusing 
- local variable: variables that are stored inside functions, they are only stored while the functions are running, and their values are stored temporarily. they are not stored in the working environment. 
- permanently alter a variable in a function, use the global then varname command before the function body. 
operations over dataframes: apply and map functions:
- create a new dataframe: data = pd.DataFrame()
- add elements to the dataframe: data["varname"] = [number1 number2 number3]
- applying functions to dataframes: apply(myfunction) takes a column vector as an input and then calculates the function seperately for each individual in the column
- creating a new variable using other variables: data['new_var'] = data['age'].apply(lambda age: age >= 18)
- dropping an existing variable in a dataframe: data = data.drop(columns=['new_var'])
- map function: executes a function for each item in an iterable list or array. 
- map syntax: list(map(function, list1,list2))
